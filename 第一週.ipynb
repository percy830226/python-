### 挑戰題一：買相機

```python
list_month = [ 'Sep', 'Oct', 'Nov', 'Dec']
list_price = [10000, 150000, 230000, 80000]
list_total =[]
price_sum = 0
```

```python
for i in list_price:
    price_sum += i
    list_total.append(price_sum)
```

```python
plt.plot(list_month, list_total, color="red", marker='.', ms=10)
```

```python

plt.pie(list_price,  labels = list_month, labeldistance = 1.1, autopct = "%3.1f%%", shadow =True, startangle = 90, pctdistance = 0.6)
```
